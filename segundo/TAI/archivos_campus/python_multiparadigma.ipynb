# Programación funcional y orientada a objetos 

En este cuaderno de jupyter haremos una introducción (opcional) a dos paradigmas distintos de programación en los que Python puede trabajar. 

## Programación funcional 

Como su nombre indica, es una forma de programar en la que el objetivo es 
aplicar funciones sobre datos (o sobre otras funciones), 
más que asignar datos a unas variables 
(programación asertiva o imperativa). 

Veremos una pequeña introducción a cómo se trabaja en este paradigma. 

### lambda 

El operador *lambda* se usa para definir funciones de una sola línea 

```python
suma = lambda x, y: x + y
print(suma(3, 5))
```

```python
# funciones anónimas: sin nombre, dentro de otro procedimiento 
a = [(1, 2), (4, 1), (9, 10), (13, -3)]
a.sort(key=lambda x: x[1])
# nótese que a la función no se le ha dado ningún nombre, de ahí el término 'anónima' 
print(a)
```

```python
# se pueden definir funciones lambda de cierta complejidad 
ab = lambda x: x if x>0 else -x
print(ab(1))
print(ab(-1))
```

### map 

La función *map* aplica una función a los elementos de un iterable 

```python
# se suele usar 'lambda' con 'map' 
# por ejemplo, en vez de este código 
items = [1, 2, 3, 4, 5]
squared = []
for i in items:
    squared.append(i**2)
squared
```

```python
# se puede hacer esto 
items = [1, 2, 3, 4, 5]
def cuadrado(x):
    return(x*x)
list(map(cuadrado, items))
```

```python
# o mejor aún esto (con una función anónima) 
list(map(lambda x: x**2, list(range(1, 6))))
```

```python
# map pude aplicarse incluso a una lista de funciones 
def multiply(x):
    return (x*x)
def add(x):
    return (x+x)
funcs = [multiply, add]
for i in range(5):
    value = list(map(lambda x: x(i), funcs))
    print(value)
```

```python
# Lo mismo en programación funcional pura: 
[list(map(lambda x: x(i), [lambda x: x*x, lambda x: x+x])) for i in range(5)]
```

### filter 

Filtra los valores de una lista según cierta condición sea verdadera o no 

```python
number_list = range(-5, 5)
less_than_zero = list(filter(lambda x: x < 0, number_list))
print(less_than_zero)
```

```python
# En una sola línea: 
list(filter(lambda x: x < 0, range(-5, 5)))
```

### reduce 

Como su nombre indica, *reduce* una lista con una función aplicada a sus elementos 

```python
# se necesita importar de la librería 'functools'  
from functools import reduce 
```

```python
# por ejemplo, en vez de este código 
producto = 1
list = [1, 2, 3, 4]
for num in list:
    producto = producto * num
producto
```

```python
# se puede usar simplemente esto 
reduce((lambda x, y: x * y), [1, 2, 3, 4])
```

A quien le guste este tipo de programación, puede buscar más información en 

https://docs.python.org/3/library/functools.html

## Programación orientada a objetos 

En este paradigma se tiene un mayor control sobre el tipo de datos que se manejan, 
sus propiedades (atributos) y los procedimientos o funciones que pueden operar sobre ellos. 
Los objetos son instancias que pertenecen a una clase (estructura de datos) 
previamente definida. Python permite también programar con clases y objetos. 

Más información se remite a la asignatura de *Programación Orientada a Objetos*. 

```python
# EJEMPLO: clase "Fraccion" para operar con fracciones (https://python.itaqui.to/intro)

class Fraccion:
    def __init__(self, numerador, denominador):
        nuevoNumerador, nuevoDenominador = simplificar(numerador, denominador)
        self.numerador = nuevoNumerador
        self.denominador = nuevoDenominador
    
    def __str__(self):
        return f"{self.numerador}/{self.denominador}"
    
    def __add__(self, otraFraccion):
        return Fraccion((self.numerador * otraFraccion.denominador) + (self.denominador * otraFraccion.numerador), (self.denominador * otraFraccion.denominador))
    
    def __sub__(self, otraFraccion):
        return Fraccion((self.numerador * otraFraccion.denominador) - (self.denominador * otraFraccion.numerador), (self.denominador * otraFraccion.denominador))
    
    def __mul__(self, otroFraccion):
        return Fraccion((self.numerador * otroFraccion.numerador), (self.denominador * otroFraccion.denominador))
    
    def __truediv__(self, otroFraccion):
        return Fraccion((self.numerador * otroFraccion.denominador), (self.denominador * otroFraccion.numerador))


def simplificar(a, b):
    from math import gcd
    d = gcd(a, b)
    return(a // d, b // d)


unMedio = Fraccion(1, 2)
unTercio = Fraccion(1, 3)
dosTercios = unTercio + unTercio

unTreceabo = Fraccion(1, 13)
unQuinceabo = Fraccion(1, 15)

print(unTercio) # Imprime "1/3"

unTercioEnTexto = str(unTercio)
print(unTercioEnTexto) # Imprime "1/3"

print(unMedio) # Imprime "1/2"
print(unMedio + unMedio) # Imprime 1/1
print(dosTercios - unTercio) # Imprime 1/3
print(unTercio * unTercio) # Imprime 1/9
print(unMedio / unMedio) # Imprime 1/1

print(unTreceabo + unQuinceabo) # Imprime "28/195"
```
