# Ficheros

```python
# Abrir un archivo de texto existente en modo lectura 'r' 
# Si el archivo no existe se producirá un error 
# Asegurarse antes de que el archivo existe antes de ejecutar esta celda! 

f = open ('archivo.txt', 'r')

# f es una variable de tipo archivo 
```

**NOTAS:** 

1. Si no se especifica el modo, se abre en modo lectura. 

2. Busca en el directorio por defecto; si está en otro lugar, 
hay que especificar el path completo donde se encuentra el archivo. 

3. Se puede cambiar el directorio actual, indicando el 'path' con un string: 

        import os
        os.chdir('path')

4. Para para prevenir problemas de codificación: 

        f = open("/path/archivo.txt", encoding="utf-8")

5. La lectura (y la escritura, que veremos después) es secuencial: 
no se puede volver a una línea anterior, ni avanzar a una línea posterior; 
se lee una por una. 

```python
# Leer una línea de un archivo de texto existente 
linea1 = f.readline()
print(linea1)
linea2 = f.readline()
print(linea2)
```

```python
# Para saber la posición actual dentro del fichero: 
f.tell()
```

```python
# Al terminar de usar el fichero, conviene cerrarlo: 
f.close()
```

```python
# Con 'with' no haría falta cerrar el fichero: 
with open('archivo.txt', 'r', encoding="utf-8") as f:
    for m in f:
        print(m, end='')
# Los elementos de un fichero de texto son sus líneas 
```

```python
# Se puede leer un fichero sin darle nombre: 
for m in open('archivo.txt', 'r', encoding="utf-8"):
    print(m)
```

```python
# Ahora generamos una lista con todas las líneas (str) del fichero: 
f = open("archivo.txt", 'r', encoding="utf-8")
L = f.readlines()
print(L)
f.close()
# En cada línea se añade automáticamente al final el carácter '\n' de fin de línea, 
# excepto posiblemente en la última línea 
```

```python
# Abrimos ahora en modo escritura 'w' 
# OJO: si existe se borran los datos al abrirlo, y si no, se crea uno nuevo 
# Conviene especificar "utf-8" para escribir acentos 
f = open ('nuevo.txt', 'w', encoding="utf-8")
```

```python
# Vamos añadiendo caracteres al final del archivo: 
f.write('\nFrase al final del fichero')
# Se devuelve el número de caracteres que se han escrito 
```

```python
# La escritura también es secuencial: cada 'write' añade texto al final del archivo 
f.write('\nOtra línea más! ')
```

```python
# IMPORTANTE: si no se cierra el fichero, no se guarda en el ordenador 
f.close()
# Ahora ir al sistema operativo y abrir el archivo correspondiente 
```

**Otros modos:** 
+ Modo añadir 'a': si existe el archivo, todos los datos se añaden al final del mismo, 
y en caso contrario se crea un archivo nuevo 
+ Modo lectura extendida 'r+' para leer y escribir indistintamente 

```python
# Volvemos al modo lectura 
f = open("archivo.txt", 'r', encoding="utf-8")
# Leemos el archivo completo: 
texto = f.read()
print(texto)
f.close()
```

```python
# Lectura mediante tamaño de buffer: indicando el número de caracteres que se quieren leer 
f = open("archivo.txt", 'r', encoding="utf-8")
texto = f.read(15)
print(texto)
f.close()
```

```python
# EJEMPLO: hacemos una copia de un archivo 
inputF = open ('archivo.txt', 'r', encoding="utf-8")
outputF = open ('copia.txt', 'w', encoding="utf-8")
N = 0
texto = inputF.read(10)
while len(texto):
    n = outputF.write(texto)
    N += n
    texto = inputF.read(10)
inputF.close()
outputF.close()
print(N, 'caracteres copiados')
```

```python
# Ficheros binarios: modos 'rb' o 'wb' 
# OJO: se leen y escriben 'bytes' 
# Comprobar que el archivo binario existe, y cambiar el nombre si procede 
inputF = open ('imagen.jpg', 'rb')
outputF = open ('copia.jpg', 'wb')
N = 0
pixels = inputF.read(100)
while len(pixels):
    n = outputF.write(pixels)
    N += n
    pixels = inputF.read(100)
inputF.close()
outputF.close()
print(N, 'bytes copiados')
```

```python
# Borrar y renombrar ficheros: 
# Funciones remove() y rename(), disponibles desde el módulo 'os' 
import os
os.remove('imagen.jpg')
os.rename('copia.jpg', 'imagen.jpg')
```

```python
# El módulo 'os' nos permite comprobar si un fichero existe: 
os.path.exists('copia.jpg')
```

## Librería 'pickle' 

Se puede usar la librería pickle para leer y escribir objetos de Python 
en ficheros binarios, por ejemplo un diccionario, una lista, etc. 

```python
# Ejemplo: 

import pickle
lista = ['perro', 'gato', 'caballo']

with open('binario.pkl', 'wb') as pickle_out:
    pickle.dump(lista, pickle_out)    # guardamos la lista en un fichero binario 

with open('binario.pkl', 'rb') as pickle_in:
    lista2 = pickle.load(pickle_in)    # importamos la lista del fichero 

print(lista2)
```

```python
# La siguiente librería es útil también para trabajar con 
# sistemas operativos de tipo Unix/Linx/Mac 

import shutil
```

shutil.copy(origen, destino) equivale al comando cp de Unix/Linux cp origen destino 
+ origen es un fichero que existe, y destino es un directorio que debe existir 
+ Si el archivo ya existe en el directorio de destino se sobreescribe 
+ Se copian también los permisos, y se actualiza la fecha de modificación 

```python
# EJEMPLO: 

# OJO: asegurarse que los directorios y los archivos existen antes de ejecutar el código 

import os
import shutil
lista_archivos = os.listdir("carpeta")
destino = "carpeta2"
for file in lista_archivos:
    # solo copiamos los archivos de texto 
    if file.endswith(".txt"):
        shutil.copy(file, destino)

# Compruése en el sistema operativo que el código ha funcionado correctamente 
```

#### Otras funciones: 

+ shutil.copyfile(origen, destino)
copia de archivo a archivo, especificando los nombres de los mismos 

    Por ejemplo: shutil.copyfile('/path/to/file', '/path/to/other/archive')


+ shutil.move(origen, destino)
mueve de forma recursiva archivo o directorio a otro lugar; 
si el destino es un directorio, todos los datos se mueven ahí, 
y si el directorio no existe se crea 

    Si en el último EJEMPLO se cambia shutil.copy por shutil.move se obtendría el resultado esperado 

Más información sobre esta librería en https://docs.python.org/3/library/shutil.html
