## Arrays de bits (bitarray) 

```python
# Este módulo maneja cadenas de bits de longitud arbitraria mediante la clase 'bitarray' 
from bitarray import bitarray
```

```python
# creamos una array de bits a partir de un bitarray vacío 
a = bitarray()

# añadimos bits al final 
a.append(True)
a.extend([False, True, True])
a
```

```python
# acceder a una componente 
a[1]
```

```python
# contar ceros y unos 
print(a.count(0))
print(a.count(1))

# longitud del bitarray 
len(a)
```

Otras formas de inicializar un array de bits:

```python
# Inicializar un bitarray con longitud dada, y bits aleatorios 
bitarray(50)
```

```python
# Definir un bitarray a partir de un string de ceros y unos 
bitarray('1001011')
```

```python
# bitarray a partir de lista, tuple o iterable de Booleanos 
lst = [True, False, False, True, False, True, True]
bitarray(lst)
```

```python
# Inicializar un bitarray de ceros con ongitud dada 
b = 20 * bitarray('0')
b
```

```python
# Otra forma de inicializar por ceros 
a = bitarray(48)
a.setall(False)
print(a)
```

```python
# slicing y reasignación 
a = 30 * bitarray('10')
a[11:37:3] = 9 * bitarray([True])
a
```

```python
# borrar una de cada tres posiciones a partir de la posición 12 
del a[12::3]
a
```

```python
# reasignación de la última posición 
a[-1:] = True
a
```

```python
# concatenar dos bitarrays 
a += bitarray('000111')
print(a)
bitarray('00') + bitarray('111')
```

```python
# se puede especificar si el primer bit es el más significativo (por defecto 'big') 
b1 = bitarray(endian='big')
```

```python
# importando desde bytes (se importa el número ascii en binario) 
b1.frombytes(b'A')
b1
```

```python
# Veamos el resultado con endian='little' 
b2 = bitarray(endian='little')
b2.frombytes(b'A')
b2
```

```python
# conversión contraria: a bytes 
b1[6] = 1
print(b1)
b1.tobytes()
```

```python
b2[1] = 1
print(b2)
b2.tobytes()
```

```python
# operaciones lógicas a nivel de bits 
# OJO si los bitarrays tienen diferente 'endian' 

a1 = bitarray('01000011')

print('NOT')
print(a1)
print(~a1)

a2 = bitarray('11010101')

print('AND')
print(a1)
print(a2)
print(a1 & a2)

print('OR')
print(a1)
print(a2)
print(a1 | a2)

print('XOR')
print(a1)
print(a2)
print(a1 ^ a2)
```

```python
# alternativa al NOT, que cambia el estado del bitarray (in place) 
print(a1)
a1.invert()
a1
```

```python
# fijar el 'endian' es esencial en conversiones (bytes y files) 
A = bitarray(endian='little')
A.frombytes(b'\x01')
A
```

```python
B = bitarray(endian='big')
B.frombytes(b'\x80')
B
```

```python
# comprobaciones: 
print(A == B)
print(A.tobytes() == B.tobytes())
```

```python
# el 'endian' no puede cambiarse una vez creado el bitarray 
# pero puede hacerse una copia cambiando la misma 
A = bitarray('11100000', endian='little')
print(A)
B = bitarray(A, endian='big')
print(B)
```

```python
# comprobaciones: 
print(A == B)
print(A.tobytes() == B.tobytes())
```

```python
# escribir bits a un archivo binario 
with open('letra.txt', 'wb') as fo:
    b1.tofile(fo)
# ir al sistema operativo y comprobar que tiene un byte: la letra C 
```

```python
# este procedimiento es alternativo para volcar un bitarray a un archivo 
with open('letra.txt', 'wb') as fo:
    fo.write(b1.tobytes())
```

**NOTAS:** 
+ En ambos métodos, si el número de bits no es múltiplo de 8 
se completa el último byte con ceros 

+ En ambos casos (fromfile/tofile) se puede especificar un segundo argumento 
con el número de bits; también completa con ceros si no es múltiplo de 8 

```python
# para leer de fichero a un bitarray hay que abrir en modo lectura 
b1 = bitarray(endian='big')
with open('letraC.txt', 'rb') as fh:
    b1.fromfile(fh)
print(b1)
```

```python
# alternativamente: 
b2 = bitarray(endian='big')
with open('letraC.txt', 'rb') as fh:
    b2.frombytes(fh.read())
print(b2)
```

```python
# codificar y decodificar a partir de un diccionario con un código instantáneo 
d = {'H':bitarray('111'), 'e':bitarray('0'),'l':bitarray('110'), 'o':bitarray('10')}
m = bitarray()
m.encode(d, 'Hello')
m
```

```python
# descodificar 
m.decode(d)
''.join(m.decode(d))
```

#### Métodos soportados por la clase bitarray 

```python
# comprueba si todas las componentes son 1 
m.all()
```

```python
# comprueba si existe algún 1 
m.any()
```

```python
# cambiar el orden de los bits 
print(m)
m.reverse()
m
```

```python
# cambiar el orden de los bits en cada byte 
M = bitarray('01000001') + bitarray('01000011')
M.bytereverse()
M
```

```python
# copiar un bitarray 
M2 = M.copy()
M2
```

```python
# saber el endianess 
M.endian()
```

```python
# rellenar con ceros al final, si la longitud no es múltiplo de 8 
# devuelve el número de bits añadidos 
m.fill()
```

```python
# búsqueda de patrones 
BB = bitarray('1100111000101111') 
BB.search(bitarray('10'))
```

```python
# buscar el primer bit coincidente con True/False 
BB.index(True)
# OJO: da error si no se encuentra 
# se puede usar find en vez de index (devuelve -1) 
```

```python
# insertar un Booleano delante de una posición 
BB.insert(0, True)
```

```python
# listado de los "atributos" del objeto 
dir(BB)
```

```python
# devuelve el bit de una posición (la última por defecto) y lo elimina del bitarray 
print(BB)
print('longitud:', len(BB))
BB.pop(3)
```

#### Utilidades 

Una sublibrería contiene varias funciones útiles para hacer conversiones con bitarrays 

```python
import bitarray.util as bu
```

```python
# Veamos todas las funciones que tiene este submódulo: 
dir(bu)
```

Con 'help' podemos averiguar cómo funcionan estas utilidades 
y qué opciones ofrecen para realizar conversiones 

Son especialmente útiles las conversiones entre bitarrays y enteros (int) 

```python
# Convertimos el entero 27 en un bitarray de 9 bits 
bu.int2ba(27, 9)
```

```python
# Recíprocamente, se puede convertir un bitarray en entero: 
bu.ba2int(bitarray('1101'))
```

```python
# comprobamos la paridad (0 si hay un número par de unos, 1 en caso contrario) 
bu.parity(BB)
```
