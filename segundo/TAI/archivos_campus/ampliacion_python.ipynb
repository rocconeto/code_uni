### Conversiones entre enteros y bytes 

```python
# crear un byte a partir de un entero con signo
i = 16
b1 = i.to_bytes(1, byteorder='big', signed=True)
print(b1)
```

```python
# crear 4 bytes a partir de un entero sin signo, empezando con el byte menos significativo
b4 = i.to_bytes(4, byteorder='little', signed=False)
print(b4)
# OJO: dentro de cada byte, el primer bit es el más significativo! 
```

```python
# bytes a partir de una lista de enteros 
bytes([255, 254, 253, 252])
```

```python
# Conversión de bytes a enteros, por defecto sin signo 
bb = b'\x10\xF0'
i = int.from_bytes(bb, byteorder='big')
print(i)
# ahora con signo 
i = int.from_bytes(b'\xF0\x0F', byteorder='big', signed=True)
print(i)
```

```python
# enteros grandes dados por una cadena de bytes, mediante lista de enteros 
i = int.from_bytes([255, 0, 0, 0], byteorder='big')
print(i)
```

```python
# Operaciones bit a bit 
byte1 = int('11110000', 2) 
byte2 = int('00001111', 2)
byte3 = int('01010101', 2)

# complemento a 1 (flip/negación bit a bit) 
# valor entero: x -> -x-1 (se interpreta como entero negativo en complemento a 2) 
# complemento a 2: -m -> |m| en binario, negación bit a bit, y sumar 1 
print('NOT:', ~byte1)

# AND, OR, XOR bit a bit 
print('AND:', byte1 & byte2)
print('OR:', byte1 | byte2)
print('XOR:', byte1 ^ byte3)

# Desplazamiento a la derecha: se pierden los bits menos significativos 
print(byte1 >> 2)
print(bin(byte1 >> 2))

# Desplazamiento a la izquierda: añade ceros a la derecha 
print(byte2 << 3)
print(bin(byte2 << 3))
```

## Diccionarios 

Un diccionario de Python es un conjunto de datos indexados por una clave, en lugar de por su posición, como sería el caso de las listas. 

Se representan mediante un conjunto, encerrado entre llaves, de parejas clave:valor. 

Las claves no pueden repetirse, pero los datos sí. 

```python
# Ejemplo de diccionario 
prefijos = {'Palencia':979, 'Segovia':921, 'Soria':975, 'Valladolid':983}
```

```python
# Añadir entradas 
prefijos['Madrid'] = 93
```

```python
# Consultar datos 
prefijos['Soria']
```

```python
# Cambiar datos 
prefijos['Madrid'] = 91
prefijos
```

```python
# Se podría haber comenzado con un diccionario vacío y añadir las entradas una por una 
agenda = {}
agenda[(9, 15)] = 'clase'
agenda[(11, 30)] = 'clase'
agenda[(13, 0)] = 'reunión'
agenda
```

**NOTA:** las claves deben ser objetos inmutables (números, strings, tuplas). 

Por tanto, las claves no pueden ser listas ni otros diccionarios (daría error). 

```python
# También se pueden generar diccionarios con la función 'dict' 
telefonos = dict(Pedro = 983000000, Juan = 921111111, Carlos = None)
telefonos
```

```python
# obtener lista de claves 
list(prefijos.keys())
```

```python
# obtener lista de valores 
list(prefijos.values())
```

```python
# obtener lista de tuplas (calve, valor) 
list(prefijos.items())
```

```python
# Se puede actualizar un diccionario con los datos de otro 
prefs = {'León':987}
prefijos.update(prefs)
prefijos
```

```python
# Ordenar los items de un diccionario según la clave o el valor 
catalogo = {'cuaderno': 4.5, 'folios': 6, 'lapicero': 1.5}

# Ordenamos los artículos por precio, de mayor a menor 
Catalogo = sorted(catalogo.items(), key = lambda x : x[1], reverse=True)
print(Catalogo)

# Ordenamos los artículos por orden alfabético de la clave 
# reverse=False es el valor por defecto, se puede omitir 
Catalogo = sorted(catalogo.items(), key = lambda x : x[0], reverse=False)
print(Catalogo)
```

### Opciones avanzadas con diccionarios 

```python
# asignar un valor por defecto a varias claves a la vez con un "iterable" 
d = {}.fromkeys('abcde', 0)
print(d)
```

```python
# comprobar si existe una clave en un diccionario 
print('a' in d)
print('f' in d)
# Nótese que los "elementos" del diccionario son las claves! 
```

```python
# pedir un dato, devolviendo un valor por defecto si la clave no existe 
d.get('a', -1)
```

```python
d.get('f', -1)
```

```python
# asignar un dato en el caso de que la clave no exista, y no cambiarlo en caso de existir 
print(d)
d.setdefault('a', 1)
d.setdefault('f', 1)
print(d)
```

```python
# diccionario a partir de una lista de tuplas 
ls = [('a', 1), ('b', 2)]
dict(ls)
```

```python
# diccionario a partir de un zip ("lista" de tuplas) 
print(list(zip('abc', range(3))))
dz = dict(zip('abc', range(3)))
print(dz)
```

```python
# zip combina dos iterables 
[x for x in zip('hola', 'majete')]
```

```python
# usar zip para actualizar con nuevas entradas clave, valor 
d.update(zip('fghi', range(2, 6)))
# el valor de 'f' se ha actualizado 
print(d)
```

**NOTA:** cuando se copia una lista o un diccionario, sobre todo si tiene iterables en sus elementos, conviene usar la "copia profunda" (from copy import deepcopy) en vez de la asignación simple, ya que entonces los objetos no son independientes, y los cambios en uno se heredan en el otro. La función **id** nos da un identificador (dirección de memoria) que nos comprueba si dos objetos son en realidad el mismo aunque tengan distinto nombre. También se puede comprobar si dos objetos x,y son el mismo con la sentencia **x is y**. La igualdad == solo comprueba si tienen el mismo valor. Si la lista solo tuviera elementos simples, bastaría con escribir por ejemplo objeto.copy().

### Control del tiempo de ejecución

Para medir el tiempo de ejecución se puede usar la librería *time* 

Habría que tomar la hora antes y después del código, e imprimir la diferencia; veamos un ejemplo:

```python
# pass no hace nada, es para que pase el tiempo, o para rellenar código por hacer 
import time
t = time.time()
for i in range(100000000):
    pass
print(time.time() - t, 'segundos')
```

**NOTA:** %%timeit en una celda de jupyter mide el tiempo de ejecución de dicha celda: 

Se ejecuta el código varias veces y se muestra una media.

```python
%%timeit
t = time.time()
for i in range(100000000):
    pass
```
